```python
#Parse JSON data from nested lists
h=[]
for i in data['list']: 
    h.append(i)
    
weather_df = pd.DataFrame(h)
```

```python
mod_df = weather_df.loc[:,['dt','main', 'weather', 'clouds', 'wind', 'visibility', 'pop', 'sys', 'dt_txt', 'rain']]
mod_df = mod_df.explode('weather').head()
```

```python
weather_data = pd.json_normalize(json.loads(mod_df.\
                                       to_json(orient="records")))
weather_data
```

```python
weather_data.columns = ['dt', 'visibility', 'pop', 'date', 'rain', 'temp', 'feels_like', 'temp_min', 'temp_max', 'pressure', 'sea_level', 'grnd_level', 'humidity', 'temp_kf', 'weather_id', 'weather_main', 'weather_description', 'weather_icon', 'clouds_all', 'wind_speed', 'wind_deg', 'wind_gust', 'sys_pod', 'rain_3h']
```
